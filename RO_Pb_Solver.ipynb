## Variables, Contraintes (1) à (6)

````python
# Restricted license - for non-production use only - expires 2024-10-28
# https://support.gurobi.com/hc/en-us/articles/4409582394769-Google-Colab-Installation-and-Licensing
````

````python
# Create model
m = Model("Solver-Tiny")

# Create variables: variables de décision
I = [i for i in range(1, nb_var + 1)]

beginning_tasks = m.addVars(I, name = "date_debut_tache", vtype = GRB.INTEGER)
for i in I: # numérotation ok car "dictionnaire"
  m.addConstr(beginning_tasks[i] >=0, "date_debut_tache" + str(i) + "positive")

machine_tasks = m.addVars(I, name = "machine_tache", vtype = GRB.INTEGER)
operator_tasks = m.addVars(I, name = "operateur_tache", vtype = GRB.INTEGER)
````

````python
# Create variables: autres variables(à déclarer)

# date fin tache i
ending_tasks = m.addVars(I, name = "date_fin_tache", vtype = GRB.INTEGER)

J = [j for j in range(1, nb_jobs +1)]
# date fin du job j
beginning_jobs = m.addVars(J, name = "date_debut_job", vtype = GRB.INTEGER)
ending_jobs = m.addVars(J, name = "date_fin_job", vtype = GRB.INTEGER)
````

````python
# Contrainte (1): date de fin tache i
for i in I:
  m.addConstr(ending_tasks[i] == beginning_tasks[i] + processing_times[i-1], "calcul_date_fin_tache" + str(i))
````

````python
# Contrainte (2): date de début du job j
for j in J:
  ist = jobs_sequence["job" + str(j)][0] # première tache de la séquence du job
  m.addConstr(beginning_jobs[j] == beginning_tasks[ist], "date_debut_job_tache" + str(j))
````

````python
# Contrainte (3): date de fin du job j
for j in J:
  ilast = jobs_sequence["job" + str(j)][-1]  # dernière tache de la séquence du job
  m.addConstr(ending_jobs[j] == ending_tasks[ilast], "date_fin_job_tache" + str(j))
````

````python
# Contrainte (4): début du job j après la release date
for j in J:
  m.addConstr(beginning_jobs[j] >= release_dates[j-1], "date_debut_job_release" + str(j))
````

````python
# Contrainte (5): respect de la séquence de tache (pour chaque job j)
for j in J:
  ilast = jobs_sequence["job" + str(j)][-1]
  for iseq in jobs_sequence["job" + str(j)]:
    if iseq < ilast:
      m.addConstr(beginning_tasks[iseq +1] >= ending_tasks[iseq], "respect_ordre_sequence" + str(j))
````

````python
# Contrainte (6): Retard et Pénalité (+ variables à créer)

tardiness_jobs = m.addVars(J, name = "retard_jobs", vtype = GRB.INTEGER)
for j in J:
  m.addConstr(tardiness_jobs[j] >= 0, "retard_positifs" + str(j))
  m.addConstr(tardiness_jobs[j] >= ending_jobs[j] - due_dates[j-1], "condition_retard" + str(j))

# Borne dont on a besoin (pour les contraintes (6) et (7))
majorant = sum(processing_times)
penalty = m.addVars(J, name = "retard_jobs", vtype = GRB.BINARY)
for j in J:
  m.addConstr(majorant*penalty[j] >= tardiness_jobs[j], "retard_positifs" + str(j))
````

## Contrainte (7)

````python
# Contrainte (7): Les taches sont effectuées sur des machines et des opérateurs dispos (i.e. une machine effectue une tache à fois, idem pour les opérateurs)
````

````python
# Encore un majorant
majorant_7 = max(tiny["parameters"]["size"]["nb_tasks"], tiny["parameters"]["size"]["nb_machines"], tiny["parameters"]["size"]["nb_operators"], majorant)
print(majorant_7)
````

````python
# Exprimons: machine utilisée pour tache i = machine utilisée pour tache i'

# matrice des machines =: carrée, taille nombre de taches, symétrique, diagonale à focer à 1 si besoin mais pas utile dans le pb, 
# coeff binaire, coeff vaut 1 si pour la tache i et tache k, même machine m
machines_egales = m.addMVar(shape = (nb_var, nb_var), name = "matrice_=_machines", vtype = GRB.BINARY)
machines_gauche = m.addMVar(shape = (nb_var, nb_var), name = "matrice_<=_machines", vtype = GRB.BINARY)
machines_droite = m.addMVar(shape = (nb_var, nb_var), name = "matrice_>=_machines", vtype = GRB.BINARY)

m.addConstrs((machines_egales[i, k] <= machines_gauche[i, k] 
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "machines_egales_gauche")

m.addConstrs((machines_egales[i, k] <= machines_droite[i, k] 
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "machines_egales_droite")

m.addConstrs((machines_egales[i, k] <= machines_droite[i, k] - (1 - machines_gauche[i, k])
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "machines_egales_droite_gauche")

# attention à la numérotation
m.addConstrs((majorant_7*machines_droite[i, k] >= machine_tasks[k+1] - machine_tasks[i+1]
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "machines_droite_remplissage_matrice")

m.addConstrs((majorant_7*machines_gauche[i, k] >= machine_tasks[i+1] - machine_tasks[k+1]
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "machines_gauche_remplissage_matrice")
````

````python
# de même pour les opérateurs

operateurs_egaux = m.addMVar(shape = (nb_var, nb_var), name = "matrice_=operateurs", vtype = GRB.BINARY)
operateurs_gauche = m.addMVar(shape = (nb_var, nb_var), name = "matrice_<=operateurs", vtype = GRB.BINARY)
operateurs_droite = m.addMVar(shape = (nb_var, nb_var), name = "matrice_>=operateurs", vtype = GRB.BINARY)

m.addConstrs((operateurs_egaux[i, k] <= operateurs_gauche[i, k] 
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "operateurs_egaux_gauche")

m.addConstrs((operateurs_egaux[i, k] <= operateurs_droite[i, k] 
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "operateurs_egaux_droite")

m.addConstrs((operateurs_egaux[i, k] <= operateurs_droite[i, k] - (1 - operateurs_gauche[i, k])
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "operateurs_egaux_droite_gauche")

# attention à la numérotation
m.addConstrs((majorant_7*operateurs_droite[i, k] >= operator_tasks[k+1] - operator_tasks[i+1]
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "operateurs_droite_remplissage_matrice")

m.addConstrs((majorant_7*machines_gauche[i, k] >= operator_tasks[i+1] - operator_tasks[k+1]
                                    for i in range(nb_var)
                                    for k in range(nb_var)
                                    if i != k), name = "operateurs_gauche_remplissage_matrice")

````

````python
def somme_matrice_gurobi(n, matrice):
  somme = 0
  for i in range(n):
    for k in range(n):
      if i!=k:
        somme += matrice[i][k]
  return somme

somme_machines_gauche = somme_matrice_gurobi(nb_var, machines_gauche)
somme_machines_droite = somme_matrice_gurobi(nb_var, machines_droite)
somme_operateurs_gauche = somme_matrice_gurobi(nb_var, operateurs_gauche)
somme_operateurs_droite = somme_matrice_gurobi(nb_var, operateurs_droite)
#print(somme_machines_gauche) # ça a l'air de renvoyer ce que je veux (en attendant que les variables soit décidées...)
somme_temps_plage_comparaison = somme_matrice_gurobi(nb_var, temps_plage_comparaison)
````

````python
# Set objective

liste_opti = [ weights[j-1]*(ending_jobs[j] + alpha*penalty[j] + beta*tardiness_jobs[j]) for j in J]

# pour que la contrainte (7) passe: coeff matrices temps_comp, droite et gauche pour opérateurs et machines
liste_7 = [somme_machines_gauche, somme_machines_droite, 
           somme_operateurs_gauche, somme_operateurs_droite,
           somme_temps_plage_comparaison]


f = m.addVar(name = "f")
m.addConstr(f ==  sum(liste_opti) + sum(liste_7), name = "fct_obj")

m.setObjective(f, GRB.MINIMIZE)
````
